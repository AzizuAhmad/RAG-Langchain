```python
### create a simple txt file

import os
os.makedirs("../data/text_files", exist_ok=True)

```

```python
sample_texts = {
    "../data/text_files/python_intro.txt": """python is a programming language
    Why Python Stands Out
    Unlike many other languages that use complex syntax and punctuation, Python uses a clean, indentation-based structure that feels almost like reading English. This makes it an ideal "first language" for those new to coding.

    Key Characteristics
    -Interpreted: You don't need to compile your code before running it. Python processes the code line-by-line, which makes debugging much faster.

    -High-Level: It abstracts away complex hardware details (like memory management), allowing you to focus entirely on solving the problem.

    -Cross-Platform: Python code written on a Mac will generally run perfectly on Windows or Linux without modification.

    -Extensive Libraries: Python has a massive "Standard Library" and thousands of third-party packages (like NumPy, Pandas, and TensorFlow) that provide pre-written code for almost any task.
    """
    ,
    "../data/text_files/ML_intro.txt":"""Building on the Python foundation, Machine Learning (ML) represents one of the most transformative applications of the language today. At its core, Machine Learning is a subset of Artificial Intelligence (AI) that focuses on building systems that learn from data to improve their performance over time, without being explicitly programmed for every specific task.

    The Paradigm Shift: Traditional vs. ML
    In traditional programming, a human writes a set of Rules and provides Data to get an Answer. In Machine Learning, we provide the Data and the Answers, and the computer creates the Rules (the model).

    The ML Workflow
    Data Collection: Gathering historical information (e.g., thousands of images of cats and dogs).

    Data Preprocessing: Cleaning and formatting the data so the machine can understand it.

    Model Training: Feeding the data into an algorithm to find patterns.

    Evaluation: Testing the model on new, unseen data to check its accuracy.

    Inference: Using the trained model to make real-world predictions.
"""
}

for filepath,content in sample_texts.items():
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)

print("files created successfully")
```

```python
### text loader
# from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader


loader = TextLoader("../data/text_files/python_intro.txt",encoding="utf-8")

docs = loader.load()
print(docs)

```

```python
### directory loader

from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding": "utf-8"},
    show_progress=False
)

documents = dir_loader.load()

print(documents)


```

```python
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf_files",
    glob="**/*.pdf",
    loader_cls=PyMuPDFLoader,
    # loader_kwargs={"encoding": "utf-8"},
    show_progress=False
)

pdf_documents = dir_loader.load()
print(pdf_documents)
```
